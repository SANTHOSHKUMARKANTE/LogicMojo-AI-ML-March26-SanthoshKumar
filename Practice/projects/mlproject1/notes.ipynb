# ✅ Corrected Version (Simple & Professional)

For example, if I clone another person’s Data Science or ML project and want to run or modify it properly, first I need to identify the `requirements.txt` file in the project. After that, I need to install the dependencies using:

```bash
pip install -r requirements.txt
```

This will install all the required libraries and packages mentioned inside the `requirements.txt` file.

Inside the `requirements.txt`, we may also have:

```text
-e .
```

This means the current project should be installed in editable mode. When pip encounters `-e .`, it triggers the `setup.py` file. The `setup.py` file helps make the project behave like a proper Python package.

Inside `setup.py`, we usually use:

```python
find_packages()
```

This function searches for folders containing `__init__.py` files. If a folder contains `__init__.py`, Python treats that folder as a package. Using this process, the project becomes properly installable and importable, and we can use clean imports throughout the project.
