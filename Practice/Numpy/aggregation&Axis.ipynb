### Aggregate functions

    sum
    mean
    min
    max
    standard devision

```python
import numpy as np

a = np.array([[1,2,3],
              [4,5,6]])

print(f"Sum of all elements: {np.sum(a)}")
print(f"Mean: {np.mean(a)}")
print(f"Maximum: {np.max(a)}")
print(f"Minimum: {np.min(a)}")
print(f"Standard Deviation: {np.std(a)}")
```

### Axis

    Axis tells NumPy which direction to aggregate
    axis = 0 => coulmn wise
    axis = 1 => row wise

```python
a = np.array([[1,2,3],
              [4,5,6]])

print(f"Sum along columns (axis=0): {np.sum(a, axis=0)}")

print(f"Sum along rows (axis=1): {np.sum(a, axis=1)}")

```
