## Operations

### Basic Operations
    a + b
    a-b
    a*b
    a/b

### Scalar Operations

    #scalar is nothing but a single value
    #vector/array (Multiple Values)
    a * 2
    a + 10

```python
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
a = np.array([[1,2,3],[4,5,6]])

print(a + 10)
print(a * 2)
```

### BROADCASTING

    👉 If shapes are different, NumPy tries to adjust automatically

🧠 CONDITION FOR BROADCASTING

    For each dimension (right → left):

    ✔ Dimensions are equal
    OR
    ✔ One of them is 1

```python
a = np.array([1,2,3])   # (3,)
b = np.array([10])     # (1,)

print(a + b)          
```

```python
#(2D + 1D)

a = np.array([[1,2,3],
              [4,5,6]])   # (2,3)

b = np.array([10,20,30])  # (3,)

print(a+b)
```

```python
#Column Broadcasting


a = np.array([[1,2,3],
              [4,5,6]])   # (2,3)

b = np.array([[10],
              [20]])      # (2,1)

print(a+b)
```

```python
#NOT COMPATIBLE → ERROR

a = np.array([1,2,3])   # (3,)
b = np.array([4,5])     # (2,)

print(a+b)
```
