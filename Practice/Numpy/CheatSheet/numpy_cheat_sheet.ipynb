# ===== Basics =====

# Numpy?

Numpy = Numerical Python

## It is used for :
    a. Fast numerical computations
    b. working with arrays (instead of lists)
    c. Backbone of ML libraries (Pandas, Scikit-learn, TensorFlow)

## Why Numpy is fast :

    a. Written in C (backend)
    b. Uses contiguous memory
    c. Supports vectorization (no loops needed)

## Single,Multi-Dimensional Arrays

    np.array()

## Special Arrays

    np.zeroes
    np.ones
    np.empty

## Range Arrays

    np.arange
    np.linspace

## Array Attributes

    arr.ndim => dimensions
    arr.shape => rows, columns
    arr.size => total elements
    a.dtype => data type
