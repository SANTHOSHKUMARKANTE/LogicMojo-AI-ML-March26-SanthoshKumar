# Numpy?

Numpy = Numerical Python

## It is used for :
    a. Fast numerical computations
    b. working with arrays (instead of lists)
    c. Backbone of ML libraries (Pandas, Scikit-learn, TensorFlow)

## Why Numpy is fast :

    a. Written in C (backend)
    b. Uses contiguous memory
    c. Supports vectorization (no loops needed)


```python
import numpy as np

a = np.array([1,2,3]) #using list to create numpy array
b = np.array((4,5,6)) #using tuple to create numpy array

print(a)
print(b)
```

## Multi-Dimensional Arrays

```python
a = np.array([[1,2,3],(4,5,6)])

a
```

## Special Arrays

    np.zeroes
    np.ones
    np.empty

```python
specialzeros = np.zeros((3,2,3)) #creates a 3D array of shape (3,2,3) filled with zeroes

specialzeros
```

```python
specialones = np.ones((3,5)) #creates a 2D array of shape (3,5) filled with ones

specialones
```

```python
specialempty = np.empty((2,3,2,1)) # creates 4D array of shape (2,3,2,1) filled with uninitialized values

# uninitialized values :: Numpy just grabs a chunk (garbage values) of memory and shows whatever values were already there.
# So the values in the array will be random and can be different each time you run the code.

specialempty
```

## Range Arrays

    np.arange
    np.linspace

```python
array = np.arange(0,10,2)

# creates a 1D array with (start,stop,step).
# start => included || stop => excluded || step => difference between each element in the array.

array
```

```python
arr = np.linspace(0,10,5)

# creates  1D Array with (start,stop,end)
# start => included || stop => included || end => number of elements in the array, and the values are equally spaced between start and stop.

arr
```

## Array Attributes

    arr.ndim => dimensions
    arr.shape => rows, columns
    arr.size => total elements
    a.dtype => data type

```python
arr = np.array([[1,2,3],(4,5,6)])

print(f"dimensions :: {arr.ndim}")
print(f"Shape :: {arr.shape}")
print(f"Size :: {arr.size}")
print(f"Data type :: {arr.dtype}")
```
