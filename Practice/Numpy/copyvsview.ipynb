## View (Shallow Copy)

    assignmenst => b = a
    slicing

```python
import numpy as np

a = np.array([1,2,3,4,5])
b = a # b is a reference to a, not a copy
print(f"Array a :: {a}")
print(f"Array b :: {b}")

b[0] = 100 # Modifying b will also modify a, since they reference the same array
print(f"Array a :: {a}")
print(f"Array b :: {b}")
```

```python
a = np.array([1,2,3,4,5])
b = a[1:4] # b is a view of a, not a copy, here b references to the same address as a, but form is different
print(f"Array a :: {a}")
print(f"Array b :: {b}")

b[0] = 100 # Modifying b will also modify a, since they reference the same array, here b[0] is a[1]
print(f"Array a :: {a}")
print(f"Array b :: {b}")
```

## Deep Copy

    b = a.copy()

```python
a = np.array([1,2,3,4,5])
b = a.copy() # b is a copy of a, not a reference, here b references to a different address than a

print(f"Array a :: {a}")
print(f"Array b :: {b}")

a[1] = 25 # Modifying a will NOT modify b, since they are different arrays
b[0] = 100 # Modifying b will NOT modify a, since they are different arrays
print(f"Array a :: {a}")
print(f"Array b :: {b}")
```
