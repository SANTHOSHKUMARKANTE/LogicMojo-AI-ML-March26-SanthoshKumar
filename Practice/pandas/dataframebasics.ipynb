### Creating a DataFrame from a List

| Input Type    | Output Shape           |
| ------------- | ---------------------- |
| `[a, b, c]`   | Column (multiple rows) |
| `[[a, b, c]]` | Row (multiple columns) |


#### The main concept behind creating single or multiple rows and columns in a DataFrame is based on the data dimension.

=> If the input is 1D (like a simple list), Pandas will treat each index as one row, resulting in a single column with multiple rows.

=> If the input is 2D (like a nested list), Pandas uses the outer structure as rows and the inner structure as columns, creating multiple columns.

👉 So, it is primarily dimension-based, and in the case of 1D data, each index is considered as a row.

```python
import pandas as pd
import numpy as np

lst = ['santhosh','hemanth','kumar','viswa','parvathi']

#column with multiple rows
df = pd.DataFrame(lst, columns=['names'])
print(df)
print()
print()

#rows with multiple columns
df = pd.DataFrame([lst])

print(df)
```

### Creating DataFrame from dict of Numpy Array

```python
data = {
    'A' : np.array([1,2,3]),
    'B' : np.array([4,5,6]),
    'C' : np.array([7,8,9])
}

df = pd.DataFrame(data)
print(df)
```

### Creating a DataFrame from a List of Dictionaries 

```python
data = [
    {'A':1,'B':2,'C':3},
    {'name':'santhosh','age':25,'city':'bangalore'},
    {'name':'hemanth','age':30,'city':'hyderabad'}
]

df = pd.DataFrame(data)

print(df)
```

### Accessing and Modifying the Index

```python
data = {
    'A' : np.array([1,2,3]),
    'B' : np.array([4,5,6]),
    'C' : np.array([7,8,9])
}

df = pd.DataFrame(data)
res = df.set_index('A')
print(res)
# print(df.index)
```
