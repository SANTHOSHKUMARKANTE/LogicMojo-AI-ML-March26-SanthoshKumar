## 2. Data Cleaning and Preprocessing

### 2.1 - Handling Mssing Values

```python
# ============================================================
# Handling Missing Values
# ============================================================
# Missing value analysis was performed on all datasets.
# Missing values were handled based on business context,
# data importance, and analytical relevance.
# Business-state null values were retained where appropriate,
# while metadata and numerical fields were imputed using
# suitable strategies.
# ============================================================



# ============================================================
# Customers Dataset
# ============================================================

customers.isna().sum()

# This dataset does not contain missing values.



# ============================================================
# Orders Dataset
# ============================================================

orders.isna().sum()

# The orders dataset contains null values in the following columns:
# - order_approved_at
# - order_delivered_carrier_date
# - order_delivered_customer_date

# These null values were analyzed using order_status to understand
# whether they represent actual data quality issues or valid
# business scenarios.

# Check order status for missing approval dates
orders[orders['order_approved_at'].isnull()]['order_status'].value_counts()

# Check order status for missing carrier delivery dates
orders[orders['order_delivered_carrier_date'].isnull()]['order_status'].value_counts()

# Check order status for missing customer delivery dates
orders[orders['order_delivered_customer_date'].isnull()]['order_status'].value_counts()

# Observation:
# Null values in approval and delivery-related columns were retained
# because they represent legitimate business situations such as:
# - canceled orders
# - unavailable orders
# - undelivered orders
# rather than corrupted or missing data.



# ============================================================
# Order Items Dataset
# ============================================================

order_items.isna().sum()

# This dataset does not contain missing values.



# ============================================================
# Payments Dataset
# ============================================================

payments.isna().sum()

# This dataset does not contain missing values.



# ============================================================
# Reviews Dataset
# ============================================================

reviews.isna().sum()

# The reviews dataset contains null values in:
# - review_comment_title
# - review_comment_message

# Observation:
# Many customers provided only numerical ratings without
# written feedback. These null values represent valid
# customer behavior rather than missing or corrupted data.

# Therefore, these null values were retained.



# ============================================================
# Products Dataset
# ============================================================

products.isna().sum()

# The products dataset contains missing values in metadata
# and product specification columns. Appropriate imputation
# techniques are applied based on business relevance and
# the nature of missing data.



# Fill missing product categories with 'unknown'
# Reason:
# Product category is important for category-level analysis.
# Assigning 'unknown' helps preserve records while identifying
# uncategorized products separately.

products['product_category_name'] = products['product_category_name'].fillna('unknown')



# Fill missing product name length with 0
# Reason:
# Missing values indicate unavailable metadata information.
# Filling with 0 avoids null-related issues during analysis.

products['product_name_lenght'] = products['product_name_lenght'].fillna(0)



# Fill missing product description length with 0
# Reason:
# Some products may not contain description information.
# Filling with 0 preserves dataset completeness.

products['product_description_lenght'] = products['product_description_lenght'].fillna(0)



# Fill missing product photos quantity with 0
# Reason:
# Missing values indicate unavailable photo information.
# Filling with 0 simplifies numerical analysis.

products['product_photos_qty'] = products['product_photos_qty'].fillna(0)



# Fill missing product weight using median
# Reason:
# Product weight is important for logistics analysis.
# Median is used because it is robust to outliers and
# preserves realistic operational values.

products['product_weight_g'] = products['product_weight_g'].fillna(
    products['product_weight_g'].median()
)



# Fill missing product length using median
# Reason:
# Product dimensions are operationally important.
# Median helps maintain realistic dimension values.

products['product_length_cm'] = products['product_length_cm'].fillna(
    products['product_length_cm'].median()
)



# Fill missing product height using median
# Reason:
# Only a very small number of records are missing.
# Median imputation safely preserves dataset consistency.

products['product_height_cm'] = products['product_height_cm'].fillna(
    products['product_height_cm'].median()
)



# Fill missing product width using median
# Reason:
# Width is important for shipping and packaging analysis.
# Median avoids distortion caused by extreme values.

products['product_width_cm'] = products['product_width_cm'].fillna(
    products['product_width_cm'].median()
)



# Verify missing values after preprocessing
products.isna().sum()



# ============================================================
# Sellers Dataset
# ============================================================

sellers.isna().sum()

# This dataset does not contain missing values.



# ============================================================
# Category Translation Dataset
# ============================================================

category_translation.isna().sum()

# This dataset does not contain missing values.



# ============================================================
# Geolocation Dataset
# ============================================================

location.isna().sum()

# This dataset does not contain missing values.



# ============================================================
# Missing Value Handling Summary
# ============================================================

# Summary:
# - Business-state null values were retained where appropriate
# - Metadata-related null values were imputed using suitable strategies
# - Numerical operational fields were filled using median values
# - No critical key columns contained missing values
# - Data integrity and business meaning were preserved during preprocessing
```
