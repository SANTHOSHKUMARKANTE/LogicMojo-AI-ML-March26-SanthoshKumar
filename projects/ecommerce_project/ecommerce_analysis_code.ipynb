## 2. Data Cleaning and Preprocessing

### 2.1 - Handling Mssing Values

```python
# ============================================================
# Handling Missing Values
# ============================================================
# Missing value analysis was performed on all datasets.
# Missing values were handled based on business context,
# data importance, and analytical relevance.
# Business-state null values were retained where appropriate,
# while metadata and numerical fields were imputed using
# suitable strategies.
# ============================================================



# ============================================================
# Customers Dataset
# ============================================================

customers.isna().sum()

# This dataset does not contain missing values.
```

```python
# ============================================================
# Orders Dataset
# ============================================================

orders.isna().sum()

# The orders dataset contains null values in the following columns:
# - order_approved_at
# - order_delivered_carrier_date
# - order_delivered_customer_date

# These null values were analyzed using order_status to understand
# whether they represent actual data quality issues or valid
# business scenarios.

# ============================================================
# Orders_approved_at Analysis
# ============================================================

# Check order status for missing approval dates
orders[orders['order_approved_at'].isnull()]['order_status'].value_counts()

# Replacing order_approved_at null values with order_purchase_timestamp for delivered orders
mask = (
    (orders['order_approved_at'].isna()) &
    (orders['order_status'] == 'delivered')
)

orders.loc[mask, 'order_approved_at'] = orders.loc[
    mask,
    'order_purchase_timestamp'
]

# print(orders[(orders['order_approved_at'].isna()) & (orders['order_status'] == 'delivered')].sum())


# ============================================================
# order_delivered_carrier_date Analysis
# ============================================================

# Check order statuses where carrier delivery date is missing
print(
    orders[
        orders['order_delivered_carrier_date'].isna()
    ]['order_status'].value_counts()
)

# -------------------------------------------------------------
# Observation:
#
# - For statuses like:
#     unavailable, canceled, processing, created, approved
#   missing carrier date is logically acceptable because
#   the order may not have reached shipping stage.
#
# - But for 'delivered' orders:
#   carrier delivery date SHOULD exist because an order
#   cannot be delivered to customer without first being
#   handed over to the carrier.
#
# - Therefore, delivered orders with missing
#   order_delivered_carrier_date are inconsistent records.
# -------------------------------------------------------------


# Display those inconsistent delivered rows
orders[
    (orders['order_delivered_carrier_date'].isna()) &
    (orders['order_status'] == 'delivered')
].head(50)


# -------------------------------------------------------------
# Create a mask for inconsistent rows:
#
# Condition:
#   1. carrier delivery date is missing
#   2. order status is delivered
#
# These rows violate normal business workflow.
# -------------------------------------------------------------
mask = (
    (orders['order_delivered_carrier_date'].isna()) &
    (orders['order_status'] == 'delivered')
)


# -------------------------------------------------------------
# Remove inconsistent rows from dataset
#
# '~mask' means:
#   keep all rows EXCEPT rows where mask=True
#
# Since only 2 rows are affected,
# dropping them is safer than imputing artificial values.
# -------------------------------------------------------------
orders = orders[~mask]




# ============================================================
# order_delivered_customer_date Analysis
# ============================================================

# Check order statuses where customer delivery date is missing
print(
    orders[
        orders['order_delivered_customer_date'].isna()
    ]['order_status'].value_counts()
)

# Display inconsistent delivered rows
orders[
    (orders['order_delivered_customer_date'].isna()) &
    (orders['order_status'] == 'delivered')
]

# Create mask for inconsistent rows
mask = (
    (orders['order_delivered_customer_date'].isna()) &
    (orders['order_status'] == 'delivered')
)

# Remove inconsistent rows
orders = orders[~mask]
```

```python
# Check missing values in reviews dataset
# print(reviews.isna().sum())

# Fill missing review titles with placeholder text
reviews['review_comment_title'] = reviews[
    'review_comment_title'
].fillna('No Title')

# Fill missing review messages with placeholder text
reviews['review_comment_message'] = reviews[
    'review_comment_message'
].fillna('No Comment')

# Verify missing values after handling
print(reviews.isna().sum())
```

```python
# Check missing values in products dataset
print(products.isna().sum())

# ---------------------------------------------------------
# Product Dataset Missing Value Analysis
#
# Group 1:
# The following columns contain 610 missing values
# for the same rows:
#
# - product_category_name
# - product_name_lenght
# - product_description_lenght
# - product_photos_qty
#
# These columns represent important product metadata.
# Since major product information is missing,
# removing these rows is a better approach.
# ---------------------------------------------------------

# Display rows with missing product metadata
# print(products[products['product_category_name'].isna()].head(50))


# Remove rows with missing core product information
products = products.dropna(
    subset=[
        'product_category_name',
        'product_name_lenght',
        'product_description_lenght',
        'product_photos_qty'
    ]
)


# ---------------------------------------------------------
# Group 2:
# The following columns contain only 2 missing values:
#
# - product_weight_g
# - product_length_cm
# - product_height_cm
# - product_width_cm
#
# These columns represent physical product dimensions.
# Instead of removing rows, we fill missing values
# using median values.
# ---------------------------------------------------------

# Display rows with missing product dimensions
# print(products[products['product_weight_g'].isnull()].head(50))


# List of dimension columns
dimension_cols = [
    'product_weight_g',
    'product_length_cm',
    'product_height_cm',
    'product_width_cm'
]


# Fill missing dimension values using median
for col in dimension_cols:
    products[col] = products[col].fillna(
        products[col].median()
    )


# Verify missing values after handling
print(products.isna().sum())
```

### 3: Data Integration (Critical Component)


```python
# Step 1 — Orders + Customers

# Merge orders with customers dataset

master_df = pd.merge(
    orders,
    customers,
    on='customer_id',
    how='left'
)
# print(orders.shape)
# print(customers.shape)
# Check shape after merge
print(master_df.shape)
# print(master_df.head())



#Step 2 — Merge Order Items

# Merge order items into master dataset

master_df = pd.merge(
    master_df,
    order_items,
    on='order_id',
    how='left'
)

# Check shape after merge
# print(order_items.shape)
print(master_df.shape)


# Step 3 — Merge Products

# Merge products dataset

master_df = pd.merge(
    master_df,
    products,
    on='product_id',
    how='left'
)

# print(products.shape)
# Check shape after merge
print(master_df.shape)

# Step 4 — Merge Payments

# Merge payments dataset

master_df = pd.merge(
    master_df,
    payments,
    on='order_id',
    how='left'
)

# print(payments.shape)
# Check shape after merge
print(master_df.shape)


#Step 5 — Merge Reviews

# Merge reviews dataset

master_df = pd.merge(
    master_df,
    reviews,
    on='order_id',
    how='left'
)

# print(reviews.shape)
# Check shape after merge
print(master_df.shape)


# Step 6 — Merge Sellers

# Merge sellers dataset

master_df = pd.merge(
    master_df,
    sellers,
    on='seller_id',
    how='left'
)

# print(sellers.shape)
# Check shape after merge
print(master_df.shape)


# Step 7 — Merge Category Translation

# Merge category translation dataset

master_df = pd.merge(
    master_df,
    category_translation,
    on='product_category_name',
    how='left'
)

# print(category_translation.shape)


# Display final master dataset

print(master_df.head())

print(master_df.info())

print(master_df.shape)
```
