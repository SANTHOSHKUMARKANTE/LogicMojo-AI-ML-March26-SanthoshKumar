Question 1:- A digital payment platform recorded the following transaction amounts made by a customer during a single day.

transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]

Write a Python program that performs the following analysis:

Identify and display all transaction values that appear more than once. Detect transactions greater than ₹5000 as these may require fraud monitoring. Calculate the total spending for the day. Create a new list containing only unique transaction values. Sort the unique transaction list in descending order to view the largest transactions first.

```python
# Given data
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]

# 1. Find values that appear more than once
repeat_values = []
for item in transactions:
    if transactions.count(item) > 1 and item not in repeat_values:
        repeat_values.append(item)
print("Values that appear more than once:", repeat_values)

# 2. Find transactions greater than 5000 (fraud monitoring)
fraud_check = []
for amount in transactions:
    if amount > 5000:
        fraud_check.append(amount)
print("Transactions above ₹5000:", fraud_check)

# 3. Calculate total spending
total_spending = 0
for amount in transactions:
    total_spending = total_spending + amount
print("Total spending for the day:", total_spending)

# 4. Create list of unique values
unique_values = []
for amount in transactions:
    if amount not in unique_values:
        unique_values.append(amount)
print("Unique transaction values:", unique_values)

# 5. Sort unique values in descending order
unique_values_desc = sorted(unique_values, reverse=True)
print("Unique values sorted (largest first):", unique_values_desc)
```

Question 2:- A dataset exported from a database contains the following customer age values.

ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46] Some entries are invalid because they are either negative numbers or missing values (None).

Write a Python program that:

Iterates through the dataset and removes all invalid age values. Stores only valid ages in a new list. Calculates the average age of valid customers. Displays the list of customers whose age is greater than 30. Prints how many invalid entries were removed during the cleaning process.

```python
# Given data
ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]

# Remove invalid ages (negative numbers and None)
valid_ages = []
invalid_count = 0

for age in ages:
    if age is not None and age >= 0:
        valid_ages.append(age)
    else:
        invalid_count = invalid_count + 1

print("Valid ages:", valid_ages)

# Calculate average age
total_age = 0
for age in valid_ages:
    total_age = total_age + age
    
average_age = total_age / len(valid_ages)
print("Average age of valid customers:", average_age)

# Find customers above 30
above_30 = []
for age in valid_ages:
    if age > 30:
        above_30.append(age)
print("Customers above 30 years old:", above_30)

# Print number of invalid entries
print("Number of invalid entries removed:", invalid_count)
```

Question 3:- An analyst wants to compare the price of products listed on two e-commerce platforms.

The following data is provided: products = ["Laptop","Phone","Tablet","Monitor","Headphones"] amazon_prices = [75000, 45000, 30000, 20000, 5000] flipkart_prices = [72000, 47000, 31000, 19500, 4800]

Write a Python program that:

Combines the product names and their prices into a single data structure representing a table. Compares the price of each product across both platforms. Determines which platform offers the lower price. Calculates the difference between the two prices. Displays the product name, cheaper platform, and price difference.

```python
# Given data
products = ["Laptop", "Phone", "Tablet", "Monitor", "Headphones"]
amazon_prices = [75000, 45000, 30000, 20000, 5000]
flipkart_prices = [72000, 47000, 31000, 19500, 4800]

# Compare each product
print("Product\t\tCheaper Platform\tPrice Difference")
print("-" * 50)

for i in range(len(products)):
    product = products[i]
    amazon = amazon_prices[i]
    flipkart = flipkart_prices[i]
    
    if amazon < flipkart:
        cheaper = "Amazon"
        difference = flipkart - amazon
    else:
        cheaper = "Flipkart"
        difference = amazon - flipkart
    
    print(f"{product}\t\t{cheaper}\t\t₹{difference}")
```

Question 4:- A web platform records the hours when users log into the system.

login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10] Write a Python program that:

Separates login hours into even-hour logins and odd-hour logins. Identifies login hours that fall into peak activity time (hours greater than 5). Sorts the login hours in ascending order. Calculates how many logins occurred during peak hours.

```python
# Given data
login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]

# Separate even and odd hours
even_hours = []
odd_hours = []

for hour in login_hours:
    if hour % 2 == 0:
        even_hours.append(hour)
    else:
        odd_hours.append(hour)

print("Even hour logins:", even_hours)
print("Odd hour logins:", odd_hours)

# Find peak hours (greater than 5)
peak_hours = []
for hour in login_hours:
    if hour > 5:
        peak_hours.append(hour)

print("Peak activity hours (after 5):", peak_hours)

# Sort login hours
sorted_hours = sorted(login_hours)
print("Sorted login hours:", sorted_hours)

# Count peak hour logins
peak_count = len(peak_hours)
print("Number of logins during peak hours:", peak_count)
```

Question 5:- An e-commerce company exports its daily order log in the following format. Each order contains multiple pieces of information:

orders = [ ["ORD001","Laptop",75000,"Completed"], ["ORD002","Mouse",500,"Completed"], ["ORD003","Laptop",75000,"Returned"], ["ORD004","Keyboard",1500,"Completed"], ["ORD005","Laptop",75000,"Completed"], ["ORD006","Monitor",20000,"Fraud"], ["ORD007","Mouse",500,"Completed"], ["ORD008","Laptop",75000,"Completed"] ]

Each row represents an order with the following structure:

[Order_ID , Product_Name , Price , Status]

Where: Order_ID → unique order identifier Product_Name → product purchased Price → price of the product Status → order status (Completed, Returned, Fraud)

Your Task:

Write a Python program to simulate a simple order analysis system using loops, nested loops, conditional statements, and list operations.

Your program must perform the following steps.

Display All Orders Using a nested loop, print the entire order table row by row so that each order’s details are displayed clearly.

Detect Suspicious Activity The system suspects fraud when: the same product appears more than 3 times, OR an order has status "Fraud"

Write a program that scans the dataset and identifies suspicious products or orders. You must use nested loops to compare product occurrences.

Calculate Total Revenue Only Completed orders contribute to revenue. Write code that: iterates through the dataset adds the price of all completed orders prints the total revenue generated for the day.

Count Order Types Using loops and conditional statements, count how many orders fall into each category:

Completed Returned Fraud Store these counts in separate variables.

Generate a List of Unique Products Create a new list containing unique product names present in the dataset. You must implement this without using the set() function.

Product Sales Analysis For each unique product, determine: how many times it was ordered total revenue generated by that product This requires nested loops.

Identify the Most Sold Product Based on the frequency of orders, determine which product appears the most in the dataset.

Remove Fraud Orders Create a cleaned dataset by removing orders whose status is "Fraud". Use list methods to store the cleaned data in a new list.

```python
# Given data
orders = [
    ["ORD001", "Laptop", 75000, "Completed"],
    ["ORD002", "Mouse", 500, "Completed"],
    ["ORD003", "Laptop", 75000, "Returned"],
    ["ORD004", "Keyboard", 1500, "Completed"],
    ["ORD005", "Laptop", 75000, "Completed"],
    ["ORD006", "Monitor", 20000, "Fraud"],
    ["ORD007", "Mouse", 500, "Completed"],
    ["ORD008", "Laptop", 75000, "Completed"]
]

# 1. Display all orders
print("=" * 60)
print("ALL ORDERS:")
print("=" * 60)
for order in orders:
    print(f"Order ID: {order[0]}, Product: {order[1]}, Price: ₹{order[2]}, Status: {order[3]}")
print()

# 2. Detect suspicious activity
# Count product occurrences
product_count = {}
for order in orders:
    product = order[1]
    if product in product_count:
        product_count[product] = product_count[product] + 1
    else:
        product_count[product] = 1

# Find suspicious products
suspicious_products = []
for product, count in product_count.items():
    if count > 3:
        suspicious_products.append(product)

# Find fraud orders
fraud_orders = []
for order in orders:
    if order[3] == "Fraud":
        fraud_orders.append(order[0])

print("Suspicious products (ordered >3 times):", suspicious_products)
print("Fraud orders:", fraud_orders)
print()

# 3. Calculate total revenue (Completed orders only)
total_revenue = 0
for order in orders:
    if order[3] == "Completed":
        total_revenue = total_revenue + order[2]
print("Total revenue from completed orders: ₹", total_revenue)
print()

# 4. Count order types
completed_count = 0
returned_count = 0
fraud_count = 0

for order in orders:
    if order[3] == "Completed":
        completed_count = completed_count + 1
    elif order[3] == "Returned":
        returned_count = returned_count + 1
    elif order[3] == "Fraud":
        fraud_count = fraud_count + 1

print("Order Statistics:")
print(f"  Completed orders: {completed_count}")
print(f"  Returned orders: {returned_count}")
print(f"  Fraud orders: {fraud_count}")
print()

# 5. Generate unique products (without using set)
unique_products = []
for order in orders:
    product = order[1]
    if product not in unique_products:
        unique_products.append(product)
print("Unique products:", unique_products)
print()

# 6. Product sales analysis
print("Product Sales Analysis:")
print("-" * 40)
for product in unique_products:
    order_count = 0
    total_sales = 0
    
    for order in orders:
        if order[1] == product:
            order_count = order_count + 1
            if order[3] == "Completed":  # Only completed orders count for revenue
                total_sales = total_sales + order[2]
    
    print(f"{product}: Ordered {order_count} times, Revenue: ₹{total_sales}")
print()

# 7. Identify most sold product
most_sold = ""
max_count = 0

for product, count in product_count.items():
    if count > max_count:
        max_count = count
        most_sold = product

print(f"Most sold product: {most_sold} (appeared {max_count} times)")
print()

# 8. Remove fraud orders
cleaned_orders = []
for order in orders:
    if order[3] != "Fraud":
        cleaned_orders.append(order)

print("Cleaned orders (fraud removed):")
for order in cleaned_orders:
    print(f"  {order[0]}: {order[1]} - ₹{order[2]} - {order[3]}")
```
