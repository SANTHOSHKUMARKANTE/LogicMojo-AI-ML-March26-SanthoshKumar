Q1. Division with Exception Handling

Write a program that takes two numbers and safely performs division. Handle division by zero using try-except.

```python
def safe_devision(num1, num2):
    try:
        res = num1 / num2
        return res
    except ZeroDivisionError:
        return "Error : Cannot divide by zero!"
print(safe_devision(10,0))
    
```

Q2. String to Integer Conversion

Given a list:

data = ["10", "20", "abc", "30"]

Convert all values to integers. Skip invalid values using exception handling.

```python
data = ["10", "20", "abc", "30"]

converted_numbers = []

for item in data:
    try:
        # Try converting the string to an integer
        number = int(item)
        converted_numbers.append(number)
    except ValueError:
        # Skip invalid values and print a warning
        print(f"Skipping invalid value: '{item}'")

print("Converted integers:", converted_numbers)
```

Q3. Multiple Exception Handling

Write a program that:

Takes user input Converts it to integer Divides 100 by that number

Handle:

ValueError ZeroDivisionError

```python
def divide_100():
    try:
        #take user input and convert it to an integer
        num = int(input("Enter a number :: "))

        #Divide 100 by the user input
        result = 100 / num

        #Display the result
        print(f"Result of 100 divide by {num} is {result}")

    except ValueError:
        print("Error: Please enter a valid number")
    except ZeroDivisionError:
        print("Error: Cannot divide by zero!")

divide_100()
```
