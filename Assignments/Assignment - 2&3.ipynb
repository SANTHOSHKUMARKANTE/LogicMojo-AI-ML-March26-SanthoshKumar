Q1. Division with Exception Handling

Write a program that takes two numbers and safely performs division. Handle division by zero using try-except.

```python
def safe_devision(num1, num2):
    try:
        # Try dividing 10 with 0
        res = num1 / num2
        return res
    except ZeroDivisionError:
        return "Error : Cannot divide by zero!"
print(safe_devision(10,0))
    
```

Q2. String to Integer Conversion

Given a list:

data = ["10", "20", "abc", "30"]

Convert all values to integers. Skip invalid values using exception handling.

```python
data = ["10", "20", "abc", "30"]

converted_numbers = []

for item in data:
    try:
        # Try converting the string to an integer
        number = int(item)
        converted_numbers.append(number)
    except ValueError:
        # Skip invalid values and print a warning
        print(f"Skipping invalid value: '{item}'")

print("Converted integers:", converted_numbers)
```

Q3. Multiple Exception Handling

Write a program that:

Takes user input Converts it to integer Divides 100 by that number

Handle:

ValueError ZeroDivisionError

```python
def divide_100():
    try:
        #take user input and convert it to an integer
        num = int(input("Enter a number :: "))

        #Divide 100 by the user input
        result = 100 / num

        #Display the result
        print(f"Result of 100 divide by {num} is {result}")

    except ValueError:
        print("Error: Please enter a valid number")
    except ZeroDivisionError:
        print("Error: Cannot divide by zero!")

divide_100()
```

Q4. Using finally

Write a program that:

Opens a file (dummy) Prints a message Ensures a “Closing file” message always runs using finally

```python
try:
    open("dummy.txt", "r")
    print("File opened")
except:
    print("Error occurred")
finally:
    print("Closing file")  # This message always runs
```

Q5. Using assert

Write a program that checks if a number is positive using assert.

```python
try:
    num = int(input("Enter a number :: "))

    assert num > 0, "Number must be positive"

    print(f"{num} is a positive number")
except AssertionError as e:
    print(f"Assertion error : {e}")
except ValueError:
    print("Invalid input! Please enter a valid number.")
```

Q12. Find Words Starting with 'a'

From:

text = "apple banana avocado berry"

Extract words starting with 'a'.

```python
text = "apple banana avocado berry"

# Normal for loop approach
result = []
for word in text.split():
    if word.startswith('a'):
        result.append(word)

result  # ['apple', 'avocado']
```

Q15. Replace Digits

Replace all digits with "X" in:

text = "My number is 98765"

```python
text = "My number is 98765"
result = ""

for i in range(len(text)):
    if text[i].isdigit():
        # If the character is a digit, replace it with 'X'
        result += "X"
    else:
        result += text[i]

result
```

Q16. Extract Dates

Extract date from:

text = "Today is 2026-03-28"

Return:

Year Month Day

```python
text = "Today is 2026-03-28"

# Simple and readable
date_str = text.split()[-1]  # Get the date part
year, month, day = date_str.split('-')  # Split into components

print(f"Year: {year}")
print(f"Month: {month}")
print(f"Day: {day}")
```

Q. 17 Extract numbers followed by "rs":

text = "100rs 200usd 300rs"

```python
text = "100rs 200usd 300rs"

# Simple loop approach
result = []
for word in text.split():
    if word.endswith('rs'):
        result.append(word[:-2])

print(result)  # ['100', '300']
```

Q. 18 Extract numbers that come after "rs":

text = "rs100 rs200 usd300"

```python
text = "rs100 rs200 usd300"

# Simple loop approach
result = []
for word in text.split():
    if word.startswith('rs'):
        result.append(word[2:])

print(result)  # ['100', '200']
```
